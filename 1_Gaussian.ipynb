```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

# 기존 분포: 평균이 1이고 표준편차가 20인 정규분포함수를 정의
mean1 = 1
std_dev1 = 20

# 새로운 분포: 평균이 4이고 표준편차가 20인 정규분포함수를 정의
mean2 = 20
std_dev2 = 20

# 추가 분포: 평균이 1이고 표준편차가 10인 정규분포함수를 정의
mean3 = 1
std_dev3 = 10

# x 값의 범위 설정
x = np.linspace(mean1 - 4*std_dev1, mean1 + 4*std_dev1, 1000)

# 정규분포 확률 밀도 함수 (PDF) 계산
pdf1 = norm.pdf(x, loc=mean1, scale=std_dev1)
pdf2 = norm.pdf(x, loc=mean2, scale=std_dev2)
pdf3 = norm.pdf(x, loc=mean3, scale=std_dev3)

# PDF 그래프 그리기
plt.plot(x, pdf1, label='Normal Distribution PDF (mean=1, std_dev=20)')
plt.plot(x, pdf2, label=f'Normal Distribution PDF (mean=4, std_dev={std_dev2})')
plt.plot(x, pdf3, label='Normal Distribution PDF (mean=1, std_dev=10)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Normal Distributions')
plt.legend()
plt.show()
```

```python
new_samples = np.random.normal(loc=0, scale=1, size=1000)
mean_value = np.mean(new_samples)
# Calculate the mean value using a for loop
sum_value = 0
for sample in new_samples:
    sum_value += sample
mean_value_loop = sum_value / len(new_samples)

# Calculate the standard deviation using a for loop
sum_squared_diff = 0
for sample in new_samples:
    sum_squared_diff += (sample - mean_value_loop) ** 2
std_deviation_loop = (sum_squared_diff / len(new_samples)) ** 0.5

# Calculate the standard deviation using numpy function
std_deviation_numpy = np.std(new_samples)

print(mean_value_loop)
print(std_deviation_loop)
print(std_deviation_numpy)
```
